```python
import matplotlib.pyplot as plt
contract_counts = df['Contract'].value_counts()
#plt.bar(contract_counts.index , contract_counts.values , color =["blue","red", "yellow"] )
plt.pie(contract_counts.values , colors = ["blue","red","yellow"] , autopct = '%1.1f%%')
plt.xlabel("Contract")
plt.ylabel("Monthlycharges")
plt.title('change in number of contract with Monthly charges')
plt.subplot()

plt.show()
```

```python
import matplotlib.pyplot as plt
contract_counts = df['Contract'].value_counts()
churn_by_contract = df.groupby('Contract')['Churn'].apply(lambda x: (x == 'Yes').mean())
print(churn_by_contract)
print(contract_counts)
plt.bar(churn_by_contract.index , churn_by_contract.values,color = ["yellow","purple","green"] )
plt.xlabel("contract type")
plt.ylabel("churn rate")
plt.title("churn rate by contract")
plt.ylim(0,1)
plt.savefig("chart.png" , dpi = 150 , bbox_inches = "tight")
plt.show()
```
